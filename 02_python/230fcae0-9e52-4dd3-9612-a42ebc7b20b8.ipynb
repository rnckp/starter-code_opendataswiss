## Description

Dieser Datensatz präsentiert die Schätzungen des jährlichen Bruttoinlandprodukts (BIP) des Kantons Freiburg und der gesamten Schweiz seit 1995.Die Daten sind sowohl in nominalen Werten (zu laufenden Preisen) als auch in realen Werten (zu Preisen des Vorjahres, d. h. preisbereinigt; Referenzjahr 2020) verfügbar, jeweils in Form von Rohwerten (na, non-adjusted) oder um die Effekte großer Sportereignisse bereinigt (nasa, sport event adjusted).Die Daten für die Schweiz stammen vom Staatssekretariat für Wirtschaft (SECO). Die Daten für den Kanton Freiburg werden vom Institut für Angewandte Wirtschaftsforschung (CREA) der Universität Lausanne im Auftrag des Amts für Statistik und Daten (SSD) des Kantons Freiburg geschätzt. Weitere Informationen zur CREA-Methodologie: [https://www.unil.ch/crea/fr/home/menuinst/indicateurs/cantonal-gdp-growth.html](https://www.unil.ch/crea/fr/home/menuinst/indicateurs/cantonal-gdp-growth.html)[](https://www.unil.ch/crea/fr/home/menuinst/indicateurs/cantonal-gdp-growth.html)[](https://www.unil.ch/crea/fr/home/menuinst/indicateurs/cantonal-gdp-growth.html)Im Rahmen einer grundlegenden Revision der Volkswirtschaftlichen Gesamtrechnungen wurden alle Zeitreihen rückwirkend im September 2025 (SECO) und im November 2025 (CREA) aktualisiert. Weitere Informationen: [https://www.seco.admin.ch/seco/fr/home/wirtschaftslage---wirtschaftspolitik/Wirtschaftslage/bip-quartalsschaetzungen-.html](https://www.seco.admin.ch/seco/fr/home/wirtschaftslage---wirtschaftspolitik/Wirtschaftslage/bip-quartalsschaetzungen-.html)

```python
# Helper function for reading datasets with the proper separator
def get_dataset(url):
    if url[-3:] != "csv":
        print("The dataset URL has no proper 'csv' extension. Reading the dataset might not have worked as expected.\nPlease check the dataset link and adjust pandas' read_csv() parameters accordingly.")
    data = pd.read_csv(url, sep=",", on_bad_lines='warn', encoding_errors='ignore', low_memory=False)
    # If the dataframe only has one column or less, the data is not comma separated; use ";" instead
    if data.shape[1] <= 1:
        data = pd.read_csv(url, sep=';', on_bad_lines='warn', encoding_errors='ignore', low_memory=False)
        if data.shape[1] <= 1:
            print("The data wasn't imported properly. Very likely the correct separator couldn't be found.\nPlease check the dataset manually and adjust the code.")
    return data
```
