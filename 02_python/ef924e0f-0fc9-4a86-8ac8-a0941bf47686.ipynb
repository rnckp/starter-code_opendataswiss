# **Bauzonen und Stand der Überbauung**

## Description

Dieser Datensatz zeigt die rechtskräftigen Bauzonenflächen nach Hauptnutzung mit ihrem Stand der Überbauung.
Datengrundlagen: Diese Statistik basiert auf den rechtskräftigen Bauzonen der kommunalen Nutzungsplanung und dem Stand der Überbauung gemäss der Übersicht über den Stand der Überbauung, der Erschliessung und der Baureife (UEB). Diese Datengrundlagen liegen in der Zuständigkeit der Gemeinde als Planungsträgerin (Art. 3 KRG). Die Flächenangaben in m² können aufgrund geometrischer Operationen und der Summierung Rundungsdifferenzen aufweisen.
Aktualität: Die Genauigkeit dieser Statistik hängt von der Aktualität der Datengrundlagen und den entsprechenden Nachführungsregeln ab. Für Bauzonen bzw. Hauptnutzungen gilt grundsätzlich eine 30-tägige Nachführungsfrist nach Inkrafttreten der Zonenfestlegung in der Nutzungsplanung (Art. 20 KGeoIV). Der Stand der Überbauung unterliegt den Nachführungsregeln der Übersicht UEB. Diese können aufgrund verschiedener Aspekte variieren, wodurch die Daten mehrere Jahre alt sein können (weitere Informationen dazu siehe Kapitel 3 der Datendokumentation Übersicht UEB und Nutzungsreserven).

```python
# Helper function for reading datasets with proper separator
def get_dataset(url):
    if url[-3:] != "csv":
        print("The dataset URL has no proper 'csv' extension. Reading the dataset might not have worked as expected.\nPlease check the dataset link and adjust pandas' read_csv() parameters accordingly.")
    data = pd.read_csv(url, sep=",", on_bad_lines='warn', encoding_errors='ignore', low_memory=False)
    # If dataframe only has one column or less the data is not comma separated, use ";" instead
    if data.shape[1] <= 1:
        data = pd.read_csv(url, sep=';', on_bad_lines='warn', encoding_errors='ignore', low_memory=False)
        if data.shape[1] <= 1:
            print("The data wasn't imported properly. Very likely the correct separator couldn't be found.\nPlease check the dataset manually and adjust the code.")
    return data
```
