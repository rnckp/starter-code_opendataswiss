# **Luftqualitätsmessungen Ostluft Thurgau – Stundenwerte**

## Description

Die hier veröffentlichten Datensätze für die Messstationen Weinfelden und Konstanz stammen aus dem automatischen Messnetz von Ostluft, das kontinuierlich verschie-dene Luftschadstoffe (z.B. NO₂, PM₁₀, O₃) und teilweise meteorologische Parame-ter (z.B. Temperatur, Luftdruck, Wind) erfasst.
Jeder Datensatz enthält zeitlich aufgelöste Einzelmesswerte im Halbstunden-, Stun-den- oder Tagesintervall, je nach Version. Die Daten geben an: Start- und Endzeit der Messung, die bezeichnende Messstelle, den Messwert, sowie den zugehöri-gen Indikator (z.B. „Luftdruck“, „NO₂“) und eine eindeutige Stations-ID (Messstation-ID).
Die Daten werden direkt aus dem standardisierten Ostluft Messnetz übernommen. Die Qualitätssicherung erfolgt gemäß den Richtlinien der Schweizer Luftreinhaltever-ordnung (LRV). Es handelt sich bei den exportierten Daten um vorläufige Werte, die erst mit der Veröffentlichung des Jahresberichts definitiv bestätigt werden. In der Regel ist die vollständige Datenbereinigung Ende April des Folgejahres abgeschlossen.
Aus Speicherplatzgründen finden sich hier nur die Daten für die letzten 365 Tage. Vollständige Datensätze für ganze Jahre können über folgenden Link als gezippte CSV heruntergeladen werden:

[https://ogdtg.ch/ostluft/stundenwert_JAHR.csv.gz](https://ogdtg.ch/ostluft/stundenwert_JAHR.csv.gz)
Beispiel 2023: [https://ogdtg.ch/ostluft/stundenwert_2023.csv.gz](https://ogdtg.ch/ostluft/stundenwert_2023.csv.gz)
Alle verfügbaren Jahre: [https://ogdtg.ch/ostluft/stundenwert_full.csv.gz](https://ogdtg.ch/ostluft/stundenwert_full.csv.gz)

Gemäss Ostluft Messkonzept wird die Messstation beim Berufsbildungszentrum in Weinfelden alle zwei Jahre (gerade Jahre; Wechselstandort), die Messstation Konstanz Wallgutstrasse alljährlich betrieben.
Die Daten können außerdem über [ostluft.ch/messwerte/datenabfrage](https://www.ostluft.ch/messwerte/datenabfrage) individuell gefiltert und als CSV-Datei heruntergeladen werden.
Weitere Informationen zu den gemessenen Schadstoffen und Indikatoren finden Sie [hier](https://www.ostluft.ch/grundlagen).
Bei Fragen wenden Sie sich an die Ostluft Datenzentrale: [datenezentrale@ostluft.ch](mailto:datenezentrale@ostluft.ch)

```python
# helper function for reading datasets with proper separator
def get_dataset(url):
    if url[-3:] != "csv":
        print("The data set URL has no proper 'csv' extension. Reading the dataset might not have worked as expected.\nPlease check the dataset link and adjust pandas' read_csv() parameters accordingly.")
    data = pd.read_csv(url, sep=",", on_bad_lines='warn', encoding_errors='ignore', low_memory=False)
    # if dataframe only has one column or less the data is not comma separated, use ";" instead
    if data.shape[1] <= 1:
        data = pd.read_csv(url, sep=';', on_bad_lines='warn', encoding_errors='ignore', low_memory=False)
        if data.shape[1] <= 1:
            print("The data wasn't imported properly. Very likely the correct separator couldn't be found.\nPlease check the dataset manually and adjust the code.")
    return data
```
