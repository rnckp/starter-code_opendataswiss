# **menuCH-Kids: Fragebogen**

## Description

Mit «menuCH-Kids» wurde erstmals eine nationale Erhebung der Ernährung von Kindern und Jugendlichen in der Schweiz durchgeführt. Die erhobenen Daten vermitteln ein umfassendes Bild ihrer Ernährungsgewohnheiten, ihres Lebensstils und ihrer gesundheitlichen Situation. 
Insgesamt 1’852 zufällig ausgewählte Mädchen (51%) und Jungen (49%) im Alter von 6 bis 17 Jahren nahmen an der Studie teil. Die Daten wurden in 6 Studienzentren, verteilt auf die drei wichtigsten Sprachregionen der Schweiz, erhoben – Bellinzona, Bern, Lausanne, Luzern, St. Gallen und Zürich.
Die Daten wurden mittels 24-Stunden-Ernährungsbefragungen (je zwei pro Person), Fragebögen zu Lebensstil und Essgewohnheiten, körperlichen Messungen sowie Blut- und Urinproben erhoben. Um die Schweizer Bevölkerung der 6 bis 17 Jährigen zu repräsentieren, wurden die erhobenen Daten statistisch gewichtet. Das heisst, die Resultate repräsentieren die Schweizer Bevölkerung von 6-17 Jahren.
